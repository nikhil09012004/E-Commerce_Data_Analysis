1. What is the total sales distribution across different categories?

Use:Barplot 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Sample - Superstore.csv", encoding='ISO-8859-1')

category_sales = df.groupby('Category')['Sales'].sum().reset_index()

sns.barplot(x='Category', y='Sales', data=category_sales, palette='pastel')
plt.title(' Sales distribution by Category')
plt.ylabel('Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Use: Pie Chart

```python

category_sales = df.groupby('Category')['Sales'].sum()

plt.figure(figsize=(6,6))
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Sales Distribution by Category')
plt.show()
```

2. Which sub-categories have the highest profit?

Use:Barplot

```python
subcategory_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False).reset_index()

sns.barplot(x='Profit', y='Sub-Category', data=subcategory_profit, palette='viridis')
plt.title('Profit by Sub-Category')
plt.xlabel('Profit ($)')
plt.tight_layout()
plt.show()
```

Use: Horizontal Dot Plot

```python
subcategory_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=True)

plt.figure(figsize=(8,6))
plt.plot(subcategory_profit.values, subcategory_profit.index, 'o', color='green')
plt.title('Profit by Sub-Category')
plt.xlabel('Profit ($)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

3. What is the monthly trend of total sales?

Use:Lineplot

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.to_period('M').astype(str) 
monthly_sales = df.groupby('Month')['Sales'].sum().reset_index()
plt.figure(figsize=(10,5))
sns.lineplot(x='Month', y='Sales', data=monthly_sales, marker='o', color='teal')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```

Use: Line Chart with markers

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.to_period('M').astype(str)

monthly_sales = df.groupby('Month')['Sales'].sum().reset_index()

plt.figure(figsize=(10,5))
plt.plot(monthly_sales['Month'], monthly_sales['Sales'], marker='o', linestyle='-', color='orange')
plt.title('Monthly Sales Trend')
plt.xticks(rotation=60)
plt.grid(True)
plt.tight_layout()
plt.show()
```

4. How do sales and profit correlate?

```python
sns.scatterplot(x='Sales', y='Profit', data=df, hue='Category')
plt.title('Sales vs Profit')
plt.tight_layout()
plt.show()
```

5. What is the sales and profit distribution across regions?

```python
region_data = df.groupby('Region')[['Sales', 'Profit']].sum().reset_index()

region_data.plot(kind='bar', x='Region', stacked=False)
plt.title('Sales and Profit by Region')
plt.ylabel('Amount ($)')
plt.tight_layout()
plt.show()
```

 Use: Grouped Box Plot

```python
region_sales_profit = df.melt(id_vars=['Region'], value_vars=['Sales', 'Profit'], var_name='Metric', value_name='Value')

plt.figure(figsize=(8,6))
sns.boxplot(x='Region', y='Value', hue='Metric', data=region_sales_profit)
plt.title('Sales & Profit Distribution by Region')
plt.yscale('log')
plt.tight_layout()
plt.show()
```

6. Which states generate the most profit? (Top 10)



Use - Barplot

```python
top_states = df.groupby('State')['Profit'].sum().sort_values(ascending=False).head(10).reset_index()

sns.barplot(x='Profit', y='State', data=top_states, palette='coolwarm')
plt.title('Top 10 States by Profit')
plt.tight_layout()
plt.show()
```

Use -Horizontal Lollipop Chart

```python
top_states = df.groupby('State')['Profit'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(8,6))
plt.hlines(y=top_states.index, xmin=0, xmax=top_states.values, color='skyblue')
plt.plot(top_states.values, top_states.index, "o")
plt.title('Top 10 States by Profit')
plt.xlabel('Profit ($)')
plt.tight_layout()
plt.show()
```

7. What is the average discount given in each category?

```python
avg_discount = df.groupby('Category')['Discount'].mean().reset_index()

sns.barplot(x='Category', y='Discount', data=avg_discount, palette='flare')
plt.title('Average Discount by Category')
plt.tight_layout()
plt.show()
```

Use-Violin Plot

```python
plt.figure(figsize=(7,5))
sns.violinplot(x='Category', y='Discount', data=df, palette='spring')
plt.title('Discount Distribution per Category')
plt.tight_layout()
plt.show()
```

8. What are the most ordered sub-categories?

Use -Barplot

```python
top_ordered = df['Sub-Category'].value_counts().reset_index()
top_ordered.columns = ['Sub-Category', 'Order Count']

sns.barplot(x='Order Count', y='Sub-Category', data=top_ordered, palette='mako')
plt.title('Most Ordered Sub-Categories')
plt.tight_layout()
plt.show()
```

9. What is the total quantity ordered per category?

```python
category_quantity = df.groupby('Category')['Quantity'].sum().reset_index()

sns.barplot(x='Category', y='Quantity', data=category_quantity, palette='Set2')
plt.title('Total Quantity by Category')
plt.tight_layout()
plt.show()
```

Use -Stripplot

```python
plt.figure(figsize=(7,5))
sns.stripplot(x='Category', y='Quantity', data=df, jitter=True, palette='muted')
plt.title(' Total Quantity per Order by Category')
plt.tight_layout()
plt.show()
```

10. Most commonly used shipping mode

Use -Barplot

```python
shipping_counts = df['Ship Mode'].value_counts().reset_index()
shipping_counts.columns = ['Ship Mode', 'Count']

sns.barplot(x='Ship Mode', y='Count', data=shipping_counts, palette='Set3')
plt.title('Preferred Shipping Modes')
plt.tight_layout()
plt.show()
```

Use: Donut Chart

```python
ship_counts = df['Ship Mode'].value_counts()
plt.figure(figsize=(7,6))
plt.pie(ship_counts, labels=ship_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'), wedgeprops=dict(width=0.6))
plt.title('Shipping Mode Distribution')
plt.show()
```
